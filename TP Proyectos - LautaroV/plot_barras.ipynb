```python
import matplotlib.pyplot as plt
```

```python
x = ['Argentina','Bolivia','Brasil','Chile','Colombia','Ecuador','Guayana Francesa','Guyana','Paraguay','Perú','Surinam','Uruguay','Venezuela']
y = [47,11,212,19,50,17,0.29,0.78,7,33,0.58,3,28]
```

```python
plt.figure(figsize=(20,6))
plt.title("Población de Sudamérica")
plt.xlabel('Paises de Sudamérica')
plt.ylabel("Habitantes (en millones)")
plt.bar(x,y, color=['cyan','green','yellow'])
plt.show()

```
