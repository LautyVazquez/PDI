```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
datos = pd.read_csv('titanic.csv')
```

```python
df1 = pd.DataFrame(datos)
```

```python
fig, ax = plt.subplots()
df1.Survived.value_counts().plot(kind = "pie", labels = ["Muertos", "Supervivientes"], title = "Distribución de supervivientes")
plt.show()
```

```python
df1.Age.plot(kind = "hist", title = "Edades ")
plt.show()
```

```python
df1.Pclass.value_counts().plot(kind = "bar", title = "Numero de personas por clase")
plt.show
```

```python
df1.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", title = "Número de muertos y supervivientes por clase")
```

```python
df1.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", stacked = True, title = "Número de personas fallecidas y supervivientes por clase")
```
